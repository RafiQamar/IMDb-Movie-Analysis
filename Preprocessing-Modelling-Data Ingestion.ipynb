```python
# here i have created a function to convert the duration values in numerical format for further analysis
# and to draw insights out of them
def duration_in_min(col):
    if pd.isna(col):
        return None
        
    col = str(col)
    
    if len(col.split(' ')) == 1:
        if 'm' in col:
            return int(col.replace('m', ''))
        elif 'h' in col:
            return int(col.replace('h', '')) * 60
            
    hour, minute = col.split(' ')
    
    if 'h' in hour and 'm' in minute:
        return int(hour.replace('h', ''))*60 + int(minute.replace('m', ''))
        
duration_in_min('2h')    
```

```python
df['Duration(min)'] = df['Duration'].apply(duration_in_min)
```

```python
# here i have created a function to convert the Number of Ratings values in numerical format for further analysis
# and to draw insights out of them
def rating_in_num(col):
    col = str(col)
    if 'K' in col:
        return float(col.replace('K', '')) * 1000     
    elif 'M' in col:
        return float(col.replace('M', '')) * 1000000
    else:    
        return float(col)
```

```python

df['Number_of_Ratings(num)'] = df['Number_of_Ratings'].apply(rating_in_num)
```

```python
# here i have created a function to convert the User Reviews values in numerical format for further analysis
# and to draw insights out of them
def user_review_num(col):
    col = str(col)
    if 'K' in col:
        return float(col.replace('K', '')) * 1000     
    else:    
        return float(col)
```

```python
df['User Reviews(num)'] = df['User Reviews'].apply(user_review_num)
```

```python
# here i have created a function to convert the Critic Reviews values in numerical format for further analysis
# and to draw insights out of them
def critic_review_num(col):
    col = str(col)
    if 'K' in col:
        return float(col.replace('K', '')) * 1000     
    else:    
        return float(col)
```

```python
df['Critic_Reviews(num)'] = df['Critic_Reviews'].apply(critic_review_num)
```

```python
'''a single movie has multiple genres, so i created a function that extract the Main genre from the movie
it will analyze a list of movie genres and match them to a predefined list of genres to find the most frequent one'''

from collections import Counter

# predefined genres that i got from imdb website
genres = ["Action", "Adventure", "Fantasy", "Drama", "Romance", 'Animation','Sci-Fi',
          'Crime', 'Comedy', 'Biography', 'Science Fiction', 'Horror', 'War', 
          'Family', 'Mystery', 'Thriller', 'Documentary', 'Musical', 'Sport', 
          'Western', 'Talk-Show', 'History', 'Game-Show', 'Music', 'News', 'Reality-TV']

def get_main_genre(movie_genres):
    
    if pd.isna(movie_genres):
        return None
    # Convert movie genres to lowercase and split by commas
    movie_genres = [genre.strip().lower() for genre in movie_genres.split(',')]
    
    # counter to keep track of matching genres
    genre_counts = Counter()
    
    # Iterate over each genre in the predefined genres list
    for predefined_genre in genres:
        normalized_predefined_genre = predefined_genre.lower()
        
        # Check how many times the predefined genre (or part of it) appears in movie genres
        for genre in movie_genres:
            if normalized_predefined_genre in genre:  # Match if part of the genre is in the movie's genre
                genre_counts[predefined_genre] += 1  # Increment the count for this predefined genre
    
    # Return the genre that matches the most
    if genre_counts:
        return genre_counts.most_common(1)[0][0]  # Get the genre with the highest count
    return None  # In case no genre matches

# Apply the function to the 'Genres' column to extract the main genre
df["Main_Genre"] = df['Genres'].apply(get_main_genre)
```

## Modelling

```python
# we have null values in Genre we can predict movie genres based on descriptions
df.loc[df.Main_Genre.isnull(), ['Main_Genre', 'Description']]
```

```python
# first I filled few missing values of description with movie name
movie_name_description = df[df.Description.isnull()]['Movie_Name']
df.loc[df['Description'].isnull(),  'Description'] = movie_name_description
df['Description'].isnull().sum()
```

__TF-IDF (Term Frequency-Inverse Document Frequency)__
* The TF-IDF vectorization process converts the movie descriptions (which are raw text data) into numerical representations that capture the importance of words in the context of the entire dataset.

* vectorizer.fit_transform takes your movie descriptions ('Description' column) and transforms them into a sparse matrix of numerical values, where each row corresponds to a movie, and each column corresponds to a unique word (term) in the entire dataset.
* The numerical value for each word represents its TF-IDF score for that particular movie description.

```python
# imported necessary libraries to create a ML model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# from xgboost import XGBClassifier
```

```python

# Vectorize the descriptions using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')

# here the Description (where null values of genre are absent), are considered as a feature.
X = vectorizer.fit_transform(df.loc[~df.Main_Genre.isnull(), 'Description'])
# genre is the target that I am trying to predict based on the description
y = df.loc[~df.Main_Genre.isnull(), 'Main_Genre']
```

```python
# split the data into training and testing set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python

# here i have trained the RandomForestClassifier model on training data
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

```python
# performance on testing set
predicted = classifier.predict(x_test)
accuracy_score(y_test, predicted)
```

We're predicting genres in a dataset with 21 different genre classes, the baseline accuracy (just guessing one class out of 21) would be 4.7%. So 45.7% would be considered much better than random guessing.

```python
# performance on training set
predicted = classifier.predict(x_train)
accuracy_score(y_train, predicted)
```

```python
# now i have trained the model on whole data
classifier = RandomForestClassifier()
classifier.fit(X, y)
```

```python
# i use the model to predict genres for the movie descriptions that do not have a genre
description_for_prediction = vectorizer.transform(df.loc[df.Main_Genre.isnull(), 'Description'])
predicted_genres = classifier.predict(description_for_prediction)
predicted_genres
```

```python
# here i inserted the predicted genres back to dataframe where Main_genre had missing values
df.loc[df.Main_Genre.isnull(), 'Main_Genre'] = predicted_genres
```
